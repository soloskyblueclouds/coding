# Creating matrix with diag values using command 'diag'. 

```python
a = np.diag([1, 2, 3])
print("Matrix with 1,2,3 in diagonal : \n" , a )

a = np.diag([1, 2, 3], 1)
print("Matrix with 1,2,3 with one shifted diagonal : \n" , a )

a = np.array([[1, 2], [3, 4]])
a1 = np.diag(a)
print("Print only diagonal elements :\n", a1)
```

# Creating matrix using command 'vander' which includes the given set of values.


```python
a = np.vander([1,2,3,4], 4)
print("Vander gives multiples of given values as below :\n", a)

a = np.vander(np.linspace(0,2,4),3)
print("Vander prints 1's, then prints values btwn 0 & 2 and then prints values \nbetween 0 & 4(which is multiple of 2):\n", a)

a = np.linspace(0,2,5)
print("Shows how linspace alone command would print: \n", a)

```

# Replicating, Joining, Mutating 





```python
a = np.array([1, 2, 3, 4, 5, 6])
b = a[:1]
print("b =", b)
b += 2
print("b =", b)
b += 1
print("b =", b)

print("\n")

a = np.array([1, 2, 3, 4, 5, 6])
b = a[:2]
print("b =", b)
b += 2
print("b =", b)
b += 1
print("b =", b)

print("\n")

a = np.array([1, 2, 3, 4, 5, 6])
b = a[:3]
print("b =", b)
b += 2
print("b =", b)
b += 1
print("b =", b)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# create some fake data
x = y = np.arange(-4.0, 4.0, 0.02)
# here are the x,y and respective z values
X, Y = np.meshgrid(x, y)
Z = np.sinc(np.sqrt(X*X+Y*Y))
# this is the value to use for the color
V = np.sin(Y)

# create the figure, add a 3d axis, set the viewing angle
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.view_init(45,60)

# here we create the surface plot, but pass V through a colormap
# to create a different color for each patch
ax.plot_surface(X, Y, Z, facecolors=cm.Oranges(V))
```
